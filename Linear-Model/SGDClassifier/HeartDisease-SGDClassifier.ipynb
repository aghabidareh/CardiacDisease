```python
import numpy as np
import pandas as pd

import random
random.seed(1441)

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import MaxAbsScaler

from sklearn.linear_model import SGDClassifier

from pickle import dump , load
```

```python
data = pd.read_csv('/content/drive/My Drive/files/heart_disease.csv')
data.head()
```

```python
x = data.iloc[: , 1:-1]
y = data.iloc[: , -1]
```

```python
scaler = StandardScaler()
mms = MinMaxScaler(clip=True , feature_range=(-1 , 1))
mas = MaxAbsScaler()

x = scaler.fit_transform(x)
x = mms.fit_transform(x)
x = mas.fit_transform(x)

x_train , x_test , y_train , y_test = train_test_split(x , y)
```

```python
params = {
    'loss' : 'perceptron',
    'penalty':'l2',
    'alpha':0.00001,
    'tol':0.0001,
    'max_iter':10000,
    'shuffle':True,
    'epsilon':0.9,
    'n_jobs' : -1,
    'verbose' : 10,
    'learning_rate':'adaptive',
    'eta0':0.001,
    'early_stopping':False,
    'warm_start':True,
    'average':True,
}

model = SGDClassifier(**params)
model.fit(x_train , y_train)
preds = model.predict(x_test)
print(classification_report(y_test , preds))
```

```python
with open('/content/drive/My Drive/models/Heart/SGDClassifier-Model.h5' , 'wb') as f:
    dump(model , f)
```
